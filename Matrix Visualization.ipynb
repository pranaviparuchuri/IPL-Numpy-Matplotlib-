# Find Highest Salaried Person

```python
h1 = Salary.max()
h1
```

```python
for i in range(10):
    for j in range(10):
        if h1 == Salary[i][j]:
            print(i,j)
```

```python
def find_keys_by_value(dic, val):
    keys = []
    for k,v in dic.items():
        if v == val:
            keys.append(k)
    return keys
person = find_keys_by_value(Pdict, 0)
year = find_keys_by_value(Sdict, 8)
print('Highest Salary person is:', person,'and his salary is', h1, 'in the year',year)
```
